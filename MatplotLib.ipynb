```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(1,10)
y=np.arange(11,20)
```

```python
a=np.arange(40,50)
b=np.arange(60,70)
```

```python
#matplot scatter
plt.scatter(x,y,c='r')
#c represents colour
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2-D scatter graph')
plt.show()
```

```python
y=x*x
```

```python
#plot library
plt.plot(x,y,'go',linestyle='dashed',linewidth=3)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2-D dashed graph')
plt.show()
```

```python
#subplot
plt.subplot(2,2,1)
plt.plot(x,y,'bo--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'r+--')
plt.show()
```

```python
#bar plot
x1=[2,3,4]
y1=[5,6,7]
x2=[10,11,12]
y2=[1,5,10]
plt.bar(x1,y1)
plt.bar(x2,y2,color='r')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('bar graph')
```

```python
plt.hist(a)
#histogram
#here y axis is the count of number
```

```python
#box plot
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True);
#vert gives vertival look
#patch_artist gives the color
```

```python
#piechart
labels=['python','C++','C','Java']
size=[200,350,500,150]
colors=['red','gold','green','pink']
explode=[0,0,0,0.4]
plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.show()
```
